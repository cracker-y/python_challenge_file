[문제]

'Account' 클래스를 작성하세요.

이 클래스는 소유자(owner)와 잔액(balance)을 속성으로 가지며, 잔액은 프라이빗 변수로 선언합니다.

'deposit'와 'withdraw' 두 개의 메소드를 통해 잔액을 조작할 수 있도록 하세요.

이 때, 잔액은 음수가 될 수 없도록 체크해야 합니다. 'Account' 객체를 생성하고,<br>
입금 및 출금 작업을 시뮬레이션한 후, 최종 잔액을 출력하세요.

```python
class Account:
    def __init__(self, owner, balance):
        self.owner = owner
        self.__balance = balance

    # 입금
    def deposit(self, gold):
        self.__balance += gold
        return gold

    # 출금
    def withdraw(self, gold):
        withdrawGold = min(self.__balance, gold)
        self.__balance -= withdrawGold
        return withdrawGold

    def __str__(self):
        return f'{self.__balance}'

# 사용자 예외 클래스
class NegativeValueError(Exception):
    def __init__(self):
        super().__init__("양수를 입력하세요.")

# 인풋 출력 함수
def getInputNumber(message, count = 0):
    return input(message).split(', ')[count]

# 인스턴스 생성
account = Account("오즈", 5000)
while True:
    try:
        # 메뉴 인풋
        selectNumber = getInputNumber("> 1.입금 <  > 2.출금 <  3.현재 잔액 (종료는 'done'을 입력하세요.)\n")
        selectNumberToInt = int(selectNumber)

        # 메뉴에 없는 번호를 입력시 continue로 실행이 더 밑으로 내려가지 않게 했습니다.
        if 3 < selectNumberToInt or 0 > selectNumberToInt:
            print()
            print("해당 메뉴의 번호를 입력 해주세요")
            print()
            continue

        while True:
            try:
                # 입금 출금 금액 input 변수 뒤로가기 사용하기 위한 변수
                valueInput = ''

                # 입금
                if 1 == selectNumberToInt:
                    # 입금 금액 입력
                    valueInput = getInputNumber("입금 금액을 입력 해주세요 > (뒤로가기 'b'을 입력하세요.)\n")
                    valueInputToInt = int(valueInput)
                    # 음수 입력시 예외 발생
                    if valueInputToInt < 0:
                        raise NegativeValueError

                    # 입금 함수
                    getDeposit = account.deposit(valueInputToInt)
                    print(f'입금 금액: {getDeposit}  [현재 잔액: {str(account)}]')
                # 출금
                if 2 == selectNumberToInt:
                    # 출금 금액 입력
                    valueInput = getInputNumber("출금 금액을 입력 해주세요 > (뒤로가기 'b'을 입력하세요.)\n")
                    valueInputToInt = int(valueInput)
                    # 음수 입력시 예외 발생
                    if valueInputToInt < 0:
                        raise NegativeValueError

                    # 출금 함수
                    getWithdraw = account.withdraw(valueInputToInt)
                    print(f'출금 금액: {getWithdraw} [현재 잔액: {str(account)}]')

                # 현재 잔액 출력
                if 3 == selectNumberToInt:
                    print(f'현재 잔액: {account}')
                print()
                break
            except ValueError:
                if 'b' == valueInput:
                    break
                print("숫자를 입력하세요!")
            except NegativeValueError as e:
                print(e)

    except ValueError:
        if 'done' == selectNumber:
            print(f'최종 잔액: {str(account)}')
            print("종료합니다.")
            break
        print("숫자를 입력하세요!")
    except NegativeValueError as e:
        print(e)
```
